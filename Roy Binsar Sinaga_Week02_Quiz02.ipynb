```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sns

data = pd.read_csv('USairpollution_quiz2.csv')
data
```

```python
#your journey & code Start here, Good luck!!!
#Plotting countplot of clusters
pal = ["#000000","#CF0A0A", "#16FF00","#FCE22A","#A8D1D1", ]
pl = sns.countplot(x=data["clusters"], palette= pal)
pl.set_title("Distribution Of The Clusters")
plt.show()
```

Berdasarkan visualiasi diatas dapat dijabarkan bahwa :
- Kluster 0 memiliki 5 anggota
- Kluster 1 memiliki 17 anggota
- Kluster 2 memiliki 2 anggota
- Kluster 3 memiliki 10 anggota
- Kluster 4 memiliki 7 anggota

Dilakukan percobaan untuk melihat hubungan antara 2 fitur pada tiap kluster menggunakan scatterplot. Setelah dilakukan 21 percobaan untuk memasangkan 2 fitur didapatkan 20 percobaan menghasilkan hubungan yang lemah bahkan cenderung tidak ada korelasi antara kedua fitur. Namun terdapat 1 percobaan yang menghasilkan hubungan yang mengarahkan ke hubungan positif dan kuat yaitu pada fitur "manu" yang merupakan jumlah perusahaan manufaktur yang mempekerjakan 20 pekerja atau lebih dan fitur "popul" yang merupakan ukuran populasi (sensus 1970); dalam ribuan. 
Pembentukan scatterplot pada pasangan "manu" dan "popul" dijelaskan pada kodingan dibawah ini.

```python
pl = sns.scatterplot(data = data,x=data["manu"], y=data["popul"],hue=data["clusters"], palette= pal)
pl.set_title("Cluster's Based On Manufactur And Population")
plt.legend()
plt.show()
```
